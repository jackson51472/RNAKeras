```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, recall_score
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
df = pd.read_csv('Churn_treino.csv', sep=';')
df
```

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
# Fazendo a padronização das colunas
# Apenas aonde não for string
stand = StandardScaler()
numerico = X.select_dtypes(include=['int64', 'float64']).columns
X[numerico] = stand.fit_transform(X[numerico])

```

```python
#Modificar as colunas Geography e Gender para numeros.
lbEncoder= LabelEncoder()
categorico = X.select_dtypes(include=['object']).columns
X[categorico] = X[categorico].apply(lbEncoder.fit_transform)
X
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Montando o formado da rede
modelo = Sequential()
modelo.add(Dense(units=120, activation='relu', input_dim=x_treinamento.shape[1]))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=120, activation='relu'))
modelo.add(Dropout(0.3))
modelo.add(Dense(units=120, activation='relu'))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modelo.fit(x_treinamento, y_treinamento, epochs=100, batch_size=32)
```

```python
previsao = modelo.predict(x_teste)
y_pred = (previsao > 0.5).astype('int32')
y_pred
```

```python
print('Acurácia: ', accuracy_score(y_teste, y_pred))
print('F1: ', f1_score(y_teste, y_pred))
print('Recall: ', recall_score(y_teste, y_pred))
print('Matriz de Confusão: ', confusion_matrix(y_teste, y_pred))
```
